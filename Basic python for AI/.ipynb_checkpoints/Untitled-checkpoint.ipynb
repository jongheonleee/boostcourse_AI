```python
# 파이썬 기초 문법

#-1. Variable & List

'''
변수란? 값을 저장하는 장소
'''

# 변수를 할당할때, 컴퓨터에서 일어나는 일?!
'''
a = 5
우항에 있는 값을 좌변에 있는 변수값에 저장해놓음

변수는 메모리 주소를 갖음
프레임
 App -> OS -> CPU | memory
 (App: 프로그래밍 언어)
* 변수는 메모리 주소를 갖고 변수에 들어가는 값은 메모리 주소에 할당됨
'''

# 컴퓨터의 구조 - 폰 노이만 아키텍쳐

'''
컴퓨터에 값을 입력/ 프로그램을 실행할 경우
-그 정보를 먼저 메모리에 저장
-CPU가 순차적으로 그 정보를 해석하고 계산하여 결과 반환
'''

# 메모리와 변수

'''
변수를 선언하는 순간, 메모리 특정영역에 물리적인 공간이 할당됨.

파이썬 -> OS -> memory
(
- 파이썬이 a라는 공간을 잡아달라고 OS한테 요청을 했고
- 그 a라는 공간에 8이라는 값을 저장을 했다.
)
'''

# 변수 이름 작명법
'''
1. 알파벳, 숫자, 언더스코어로 선언 가능
2. 의미 있는 단어로 표기
3. 대소문자가 구분
4. 예약어(for, if, else ,,,)는 사용안함
'''

# Basic Operation
'''
> 1. 기본 자료형
> 2. 연산자와 피연산자
> 3. 데이터 형변환
'''
#1. 수치, 문자, 논리/불린 자료형
'''
데이타 타입에 따라 메모리의 크기가 다름
정수:32bit
실수:4byte,,,
'''

a = 1
b = 1.5
print(a, b)
print(type(a))
print(type(b))

c1 = 'absiwp;ndw'
c2 = 'ldanwndwac'

print(type(c1))
print(type(c2))

t = True
f = False
print(t)
print(f)
print(type(t))


```

```python
# 이차원 리스트 카피
# 복사가 일어나는 구문[:]은 원 디멘셔널에선 적용되는 데 투 디멘셜에선 적용안됨
# copy 라이브러리 활용
import copy

matrix = [[i for i in range(1, 10) for _ in range(5)]]
copy_matrix = copy.deepcopy(matrix)

matrix[0][3] = 20

print(matrix[0][3])
print(copy_matrix[0][3])
```
